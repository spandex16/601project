How do different types of Pokémon correlate with their overall strength, as indicated by the 'Total' attribute?(Spandan)

The total attribute factor is just a simple addition of all of its other attributes. Just by looking at the data there is no way to tell if there is a specific attribute that contributes more to this total value on average and why that might determine if a pokemon is stronger. The addition of the adjacency matrix will show which types are stronger on average and then it can be correlated with the total attribute.


What insights can be gained from analyzing the relationship between specific attributes (e.g., Attack, Defense, Speed) and a Pokémon's effectiveness in battles?(Spandan)

In order to determine this a rating is called the combat power which is used by the game developers of Pokemon Go. It is a value assigned to each pokemon to determine its performance in battle by using the weighted value of its attributes. This will allow us to find a more normalized value for all the pokemon in comparison to its ‘Total Attribute’ value and give us a better understanding as to which pokemon is better.

```python
def Combat_Power(attack,special_atk,defense,special_def,hp):
    total_a = attack + special_atk
    total_d = defense + special_def
    numerator = math.floor(total_a * (total_d**0.5)*(hp**0.5)*(0.667934**2)) # Taking the average level as 25 so used the cpm value for level 25
    return math.floor(max(10,numerator/10))

#https://pokemongohub.net/post/wiki/cp-mechanics/
#https://pokemongohub.net/post/article/pokemon-go-cpm-list/
```

```python
data['CP'] = data.apply(lambda row: Combat_Power(row['Attack'], row['Sp. Atk'], row['Defense'], row['Sp. Def'], row['HP']), axis=1)

data.head()
```

```python
generation_cp = data.groupby('Generation')['CP'].mean()
plt.figure(figsize=(10, 6))
plt.plot(generation_cp.index, generation_cp.values, marker='o', linestyle='-')
plt.title('Average CP Changes with Generation')
plt.xlabel('Generation')
plt.ylabel('Average CP')
plt.grid(True)
plt.show()
```

```python
def update_plot(type_selected, evolution_selected, generation_selected):
    filtered_data = data[
        (data['Type 1'] == type_selected) | (data['Type 2'] == type_selected) &
        (data['Evolution'] == evolution_selected) &
        (data['Generation'] == generation_selected)
    ]


    if not filtered_data.empty:
        plt.figure(figsize=(12, 8))
        sns.boxplot(x='CP', y='Type 1', data=filtered_data, dodge=True)
        plt.title(f'Box Plot of CP for {type_selected} Type Pokémon\n'
                  f'Evolution: {evolution_selected}, Generation: {generation_selected}')
        plt.xlabel('CP')
        plt.ylabel('Type')
        plt.grid(True)
        plt.show()
    else:
        print('No Pokemon available for the selected filters.')


all_types = pd.concat([data['Type 1'], data['Type 2']]).unique()
all_evolution = data['Evolution'].unique()
all_generations = data['Generation'].unique()


type_dropdown = widgets.Dropdown(options=['All'] + list(all_types), value='All', description='Select Type:')
evolution_dropdown = widgets.Dropdown(options=['All'] + list(all_evolution), value='All', description='Evolution:')
generation_dropdown = widgets.Dropdown(options=['All'] + list(all_generations), value='All', description='Generation:')


interactive_plot = interactive(
    update_plot,
    type_selected=type_dropdown,
    evolution_selected=evolution_dropdown,
    generation_selected=generation_dropdown
)


display(interactive_plot)
```

```python
def update_plot(type_selected, evolution_selected):
    filtered_data = data[
        (data['Type 1'] == type_selected) | (data['Type 2'] == type_selected) &
        (data['Evolution'] == evolution_selected)
    ]

    
    if not filtered_data.empty:
        plt.figure(figsize=(12, 8))
        sns.kdeplot(x='CP', hue='Generation', data=filtered_data, fill=True, common_norm=False)
        plt.title(f'Ridge Plot of CP for {type_selected} Type Pokémon\n'
                  f'Evolution: {evolution_selected}')
        plt.xlabel('CP')
        plt.ylabel('Density')
        plt.grid(True)
        plt.show()
    else:
        print('No Pokemon available for the selected filters.')


all_types = pd.concat([data['Type 1'], data['Type 2']]).unique()
all_evolution = data['Evolution'].unique()


type_dropdown = widgets.Dropdown(options=['All'] + list(all_types), value='All', description='Select Type:')
evolution_dropdown = widgets.Dropdown(options=['All'] + list(all_evolution), value='All', description='Evolution:')


interactive_plot = interactive(
    update_plot,
    type_selected=type_dropdown,
    evolution_selected=evolution_dropdown,
)


display(interactive_plot)
```
